## 1. Helper functions

### 1.1 Helper function to fetch H3 id for lat/lon

```python
def safe_h3_index(lat, lng, res=9):
    # Check if lat/lng are valid numbers
    if pd.isna(lat) or pd.isna(lng):
        return None
    # Validate coordinate range
    if not (-90 <= lat <= 90 and -180 <= lng <= 180):
        return None
    try:
        return h3.latlng_to_cell(lat, lng, res)
    except:
        return None
```

## 2. Read Initial data set into data frame and keep only required columns

## 6. Update CD_CATEG_ROUTE &  VITESSE_AUTOR with majority rule
###. Drop the columns LOC_LAT, LOC_LONG, CD_CATEG_ROUTE, and DD_ACCDN, then remove any rows containing null values.

```python
file_path = 'accnd_data_with_weather_Info.csv'
df = pd.read_csv(file_path)

# Columns to drop
cols_to_drop = ['LOC_LAT', 'LOC_LONG', 'DT_ACCDN', 'DD_ACCDN','CD_CATEG_ROUTE_']

# Safety check: only drop columns that actually exist
existing_cols_to_drop = [c for c in cols_to_drop if c in df.columns]
missing_cols = [c for c in cols_to_drop if c not in df.columns]

if missing_cols:
    print(f'Warning: columns not found and skipped: {missing_cols}')

# Drop specified columns
df_edited = df.drop(columns=existing_cols_to_drop)

# Drop any rows with null values in any column
rows_before = len(df_edited)
df_edited = df_edited.dropna()

rows_dropped = rows_before - len(df_edited)

print(f'Columns dropped    : {existing_cols_to_drop}')
print(f'Rows removed (null): {rows_dropped:,}')
print(f'df_edited shape    : {df_edited.shape}')
print(f'Remaining nulls    : {df_edited.isnull().sum().sum()}')

print(df_edited.shape)
df_edited.head(2)
```

## 6.1: Create `df_majority_rules`

Group `df_edited` on `ZONE`, `DT_ACCDN`, and `HEURE_ACCDN`. For each group:
- `NB_VICTIMES_ZONEHR` = **sum** of `NB_VICTIMES_TOTAL_PER_ZONE_MM_DD_HH` across the group
- All other columns = **most frequent value** in the group (ties broken at random)

```python
def majority_vote(series):
    """Return the most frequent value in a Series. Break ties randomly."""
    value_counts = series.value_counts()
    max_count = value_counts.iloc[0]
    # Collect all values tied for the highest frequency
    tied_values = value_counts[value_counts == max_count].index.tolist()
    return np.random.choice(tied_values)


# Grouping columns
group_cols = ['ZONE', 'MM_ACCDN', 'JR_SEMN_ACCDN', 'HEURE_ACCDN']

# Column to sum
sum_col = 'NB_VICTIMES_TOTAL_PER_ZONE_MM_DD_HH'

# All remaining columns get majority vote
majority_cols = [c for c in df_edited.columns if c not in group_cols and c != sum_col]

print(f'Grouping on   : {group_cols}')
print(f'Sum column    : {sum_col} -> NB_VICTIMES_ZONEHR')
print(f'Majority vote : {majority_cols}')
```

```python
# Build aggregation dictionary
agg_dict = {sum_col: 'sum'}
for col in majority_cols:
    agg_dict[col] = majority_vote

# Group and aggregate
df_majority_rules = (
    df_edited
    .groupby(group_cols, as_index=False)
    .agg(agg_dict)
)

# Rename the summed victims column
df_majority_rules = df_majority_rules.rename(columns={sum_col: 'NB_VICTIMES_ZONEHR'})

# Reorder: group cols first, then NB_VICTIMES_ZONEHR, then the rest
remaining_cols = [c for c in df_majority_rules.columns if c not in group_cols + ['NB_VICTIMES_ZONEHR']]
df_majority_rules = df_majority_rules[group_cols + ['NB_VICTIMES_ZONEHR'] + remaining_cols]

df_majority_rules.to_csv('Final_Training_DataSet_v3.csv', index=False)      

df_majority_rules.head(10)
```
