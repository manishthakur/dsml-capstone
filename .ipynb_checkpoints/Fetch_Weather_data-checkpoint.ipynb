## 1. Helper Functions

### Function to fetch Historic weather info from api.weather.gc.ca

```python
def get_canada_weather(station_id, date_time, interval='hourly'):

    time.sleep(0.001)
    """
    Fetches weather data from api.weather.gc.ca.
    Args:
        station_id (str): The Climate Station ID (e.g., '702S006' for Montreal)
        date_time (str): The date in 'YYYY-MM-DD' or 'YYYY-MM-DDTHH:MM:SSZ' format.
        interval (str): 'hourly' or 'daily' observations.
    Returns:
        dict: The JSON response containing weather properties.
    """
    # Determine the collection based on interval
    collection = "climate-hourly" if interval == 'hourly' else "climate-daily"

    # Base URL for the ECCC OGC API
    base_url = f"https://api.weather.gc.ca/collections/{collection}/items"

    # print('fetching data for station ' + station_id +' and time ' + date_time)
    # Parameters for the API call
    params = {
        "f": "json",               # Format
        "CLIMATE_IDENTIFIER": station_id,
        "datetime": date_time,      # Can be a single date or a range like '2023-01-01/2023-01-02'
    }
    try:
        response = requests.get(base_url, params=params)
        response.raise_for_status() # Raise error for bad status codes
        data = response.json()
        # print(data)
        # Simplify the output into a list of weather properties
        features = data.get('features', [])
        return [f['properties'] for f in features]

    except requests.exceptions.RequestException as e:
        print(f"Error fetching data: {e}")
        return None

# --- TEST USAGE OF  get_canada_weather ---
# Fetching hourly data for Montreal (Station 7025250) for a specific day
weather_data = get_canada_weather(station_id='6158359', date_time='2025-01-15T14:00:00Z')
if weather_data:
    df_weather = pd.DataFrame(weather_data)
    print(df_weather[['PRECIP_AMOUNT', 'TEMP', 'WIND_SPEED']].head())
```

### Function to convert Date and Hour string into required Timestamp format for API call

```python
def get_timestamp(date_str, heure_val):
    """
    date_str: string like 'YYYY-MM-DD'
    heure_val: string like '02:00:00-02:59:00'
    Returns 'YYYY-MM-DDTHH:MM:SSZ' or '' if missing.
    """
    if heure_val is None or heure_val == 'Non précisé':
        return ''

    final_date_time_str = ''
    if pd.notna(heure_val):
        start_time = str(heure_val).split('-')[0]
    else:
        start_time = None

    if start_time:
        final_date_time_str = f"{date_str}T{start_time}Z"
    return final_date_time_str
```

### Function to fetch nearest weather station by providing row with lat & lon

```python
def get_station_id_for_row(row):
    """
    Helper to extract lat/long from a row and return station ID.
    """
    lat = row['LOC_LAT']
    lon = row['LOC_LONG']

    # Only call API if coordinates are valid
    if pd.notnull(lat) and pd.notnull(lon):
        station = get_nearest_station(lat, lon)
        # Optional: Add a tiny delay to be polite to the API
        time.sleep(0.1)
        return station
    return None
```

### Fuction to fetch nearest weather station id by lat and lon

```python
def get_nearest_station(lat, lon):
    """
    Finds the nearest weather station to a given coordinate using the ECCC API.
    """
    # 1. Define a small bounding box (approx 10km radius) around the point
    # Longitude is first in OGC API [min_lon, min_lat, max_lon, max_lat]
    buffer = 0.1
    bbox = f"{lon-buffer},{lat-buffer},{lon+buffer},{lat+buffer}"
    print(bbox)

    url = "https://api.weather.gc.ca/collections/climate-stations/items"
    params = {
        "f": "json",
        "bbox": bbox,
        "limit": 5  # Get a few candidates to find the absolute closest
    }

    try:
        response = requests.get(url, params=params)
        response.raise_for_status()
        data = response.json()

        stations = data.get('features', [])
        if not stations:
            return None

        # 2. Simple distance calculation to find the best match
        best_station = None
        min_dist = float('inf')

        for s in stations:
            s_lon, s_lat = s['geometry']['coordinates']
            # Using Euclidean distance (fine for small local distances)
            dist = ((lat - s_lat)**2 + (lon - s_lon)**2)**0.5

            if dist < min_dist:
                min_dist = dist
                best_station_id = s['properties']['CLIMATE_IDENTIFIER']
                best_station_name = s['properties']['STATION_NAME']

        print('lat / lon ' + str(lat) + str(lon) + ' ' + str(best_station_id) + ' : '+ str(best_station_name))
        return best_station_id

    except Exception as e:
        print(f"Error finding station: {e}")
        return None

# --- Example Usage ---

lat, lon = 45.455505, -73.861616
station_id = get_nearest_station(lat, lon)


print(f"The nearest station ID for this accident is: {station_id}")
```

### Read Group Accidental Data

```python
# --- Load and subset ---
df = pd.read_csv(initia_data_set)

all_selected_features_for_weather = ['LOC_LAT','LOC_LONG', 'DT_ACCDN', 'HEURE_ACCDN']

df_subset = df[all_selected_features_for_weather].copy()

# Ensure DT_ACCDN is datetime and create date string
df_subset['DT_ACCDN'] = pd.to_datetime(df_subset['DT_ACCDN'], errors='coerce')
df_subset['DT_ACCDN_STR'] = df_subset['DT_ACCDN'].dt.strftime('%Y-%m-%d')

# Setting same id for whole montreal island to fetch the historic weather infomation
df_subset.insert(0, 'STATION_ID', '702S006')

df_subset['DD_ACCDN'] = df_subset['DT_ACCDN'].dt.day

# You need to pass the row to the function
df_subset['HEURE_ACCDN_FOR_WEATHER_API'] = df_subset.apply(lambda row: get_timestamp(row['DT_ACCDN_STR'], row['HEURE_ACCDN']), axis=1)

df_subset = df_subset.filter(['STATION_ID', 'HEURE_ACCDN_FOR_WEATHER_API'])

df_subset.head(5)
```

### Create a unique pair of STATION_ID and HEURE_ACCDN_FOR_WEATHER_API to avoid duplicate weather API calls

```python
## Create a unique weather dataframe with unique (station_id and date_time) to fetch weather data

weather_request_subset_df = df_subset.copy()
print('Total STATION_ID & HEURE_ACCDN_FOR_WEATHER_API combination : '+ str(len(weather_request_subset_df)))

for col in ['STATION_ID', 'HEURE_ACCDN_FOR_WEATHER_API']:
    weather_request_subset_df[col] = (
        weather_request_subset_df[col].astype('string').str.strip().replace('', pd.NA)
    )

weather_request_subset_df = weather_request_subset_df.dropna(subset=['STATION_ID', 'HEURE_ACCDN_FOR_WEATHER_API'])
weather_request_subset_df = weather_request_subset_df.drop_duplicates(subset=['STATION_ID', 'HEURE_ACCDN_FOR_WEATHER_API']).reset_index(drop=True)

print('Total Unique combination : '+ str(len(weather_request_subset_df)))
```

### Fetch Weather Data


```python
# 1. FIX: Assign the first 50 rows of your DataFrame to this variable
# Do NOT use unique_station_dates_limit = []
unique_station_dates_limit = weather_request_subset_df.head(50)

# 2. Create a SEPARATE list to hold the incoming API results
all_weather_records = []

# 3. Fetch weather data in batches
for index, row in unique_station_dates_limit.iterrows():
    station_id = row['STATION_ID']
    date_str = row['HEURE_ACCDN_FOR_WEATHER_API']
    
    weather_for_day = get_canada_weather(station_id=station_id, date_time=date_str, interval='hourly')
    
    if weather_for_day:
        for record in weather_for_day:
            record['STATION_ID'] = station_id
            record['HEURE_ACCDN_FOR_WEATHER_API'] = date_str
            # 4. Append to the results list, NOT the dataframe you are iterating over
            all_weather_records.append(record)

    time.sleep(0.1)

# 5. Convert the results list into a new DataFrame once the loop is done
weather_results_df = pd.DataFrame(all_weather_records)
weather_results_df.head()

selected_cols = ['CLIMATE_IDENTIFIER', 'HEURE_ACCDN_FOR_WEATHER_API', 'PRECIP_AMOUNT', 'TEMP', 'WIND_SPEED']
df_all_fetched_weather_records_subset = weather_results_df[selected_cols]

df_all_fetched_weather_records_subset.to_csv('combined_weather_data_702S006.csv', index=False)
df_all_fetched_weather_records_subset.head()
```
