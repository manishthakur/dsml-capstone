# Risk Classification Notebook — Tweedie × XGBoost Grid

## 0 · Imports

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import TweedieRegressor
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from xgboost import XGBClassifier

RANDOM_STATE = 42
TWEEDIE_CLIP = 1e-3   # NB_VICTIMES_ZONEHR is clipped to this minimum when power > 1
```

## 1 · Load Data

```python
df_raw = pd.read_csv('Final_Training_DataSet_v3.csv')
df_raw.head()
```

## 2 · Drop `CD_CATEG_ROUTE`

```python
# df_base = df_raw.drop(columns=['CD_CATEG_ROUTE'])
df_base = df_raw.drop(columns=['CD_CATEG_ROUTE'])
df_base.head()
```

## 3 · Create RISK_CLASS Versions

In all versions: `NB_VICTIMES_ZONEHR == 0` → `A`

| Version | B | C | D |
|---------|---|---|---|
| **a** | ==1 | >1 | — |
| **b** | 1–2 | >2 | — |
| **c** | ==1 | ==2 | >2 |
| **d** | 1–2 | ==3 | >3 |
| **e** | 1–2 | 3–4 | >4 |

```python
def mapping_a(v):
    if v == 0: return 'A'
    if v == 1: return 'B'
    return 'C'

def mapping_b(v):
    if v == 0:      return 'A'
    if v in (1, 2): return 'B'
    return 'C'

def mapping_c(v):
    if v == 0: return 'A'
    if v == 1: return 'B'
    if v == 2: return 'C'
    return 'D'

def mapping_d(v):
    if v == 0:      return 'A'
    if v in (1, 2): return 'B'
    if v == 3:      return 'C'
    return 'D'

def mapping_e(v):
    if v == 0:      return 'A'
    if v in (1, 2): return 'B'
    if v in (3, 4): return 'C'
    return 'D'

versions = {}
for name, fn in [('a', mapping_a), ('b', mapping_b), ('c', mapping_c),
                 ('d', mapping_d), ('e', mapping_e)]:
    df_v = df_base.copy()
    df_v['RISK_CLASS'] = df_v['NB_VICTIMES_ZONEHR'].apply(fn)
    versions[name] = df_v

for name, df_v in versions.items():
    print(f'Version {name}:')
    print(df_v['RISK_CLASS'].value_counts().sort_index(), '\n')
```

## 4 · Preprocessing

```python
DAY_ORDER = ['LU', 'MA', 'ME', 'JE', 'VE', 'SA', 'DI']
day_map = {d: i for i, d in enumerate(DAY_ORDER)}

zone_le = LabelEncoder()
zone_le.fit(df_base['ZONE'])
```

## 5 · Custom Severity Score

```python
def severity_score(y_true: np.ndarray, y_pred: np.ndarray) -> int:
    """
    - Correct prediction  : +1
    - Under-prediction    : -(actual_idx - predicted_idx)
    - Over-prediction     :  0
    """
    score = 0
    for actual, predicted in zip(y_true, y_pred):
        if actual == predicted:
            score += 1
        elif actual > predicted:
            score -= (actual - predicted)
    return score
```

## 6 · Tweedie Variants

10 variants spanning different `power` values and regularisation strengths.  
`NB_VICTIMES_ZONEHR` is clipped to `TWEEDIE_CLIP` (1e-3) whenever `power > 1` to satisfy the strictly-positive target requirement.

```python
# Each entry: (label, power, alpha, max_iter)
# power=0   → Gaussian (no clipping needed)
# power=1   → Poisson  (no clipping needed)
# power>1   → compound Poisson-Gamma / Gamma → requires strictly positive target
TWEEDIE_VARIANTS = [
    ('TW-01', 0.0,  0.1,  300),   # Gaussian,               low regularisation
    ('TW-02', 0.0,  1.0,  300),   # Gaussian,               medium regularisation
    ('TW-03', 1.0,  0.1,  300),   # Poisson,                low regularisation
    ('TW-04', 1.0,  1.0,  300),   # Poisson,                medium regularisation
    ('TW-05', 1.0,  5.0,  300),   # Poisson,                high regularisation
    ('TW-06', 1.5,  0.1,  500),   # compound Poisson-Gamma, low regularisation
    ('TW-07', 1.5,  0.5,  500),   # compound Poisson-Gamma, medium regularisation
    ('TW-08', 1.5,  2.0,  500),   # compound Poisson-Gamma, high regularisation
    ('TW-09', 1.9,  0.5,  500),   # near-Gamma,             medium regularisation
    ('TW-10', 2.0,  0.5,  500),   # Gamma,                  medium regularisation
]

print(f'{len(TWEEDIE_VARIANTS)} Tweedie variants defined.')
```

## 7 · XGBoost Variants

10 variants exploring learning rate, depth, subsampling, and regularisation.

```python
# Each entry: (label, n_estimators, max_depth, learning_rate, subsample,
#              colsample_bytree, reg_alpha, reg_lambda, tweedie_variance_power)
XGB_VARIANTS = [
    ('XGB-01', 200, 4, 0.10, 0.8, 0.8, 0.0, 1.0, 1.5),   # shallow,  moderate lr
    ('XGB-02', 200, 6, 0.10, 0.8, 0.8, 0.0, 1.0, 1.5),   # medium,   moderate lr
    ('XGB-03', 300, 6, 0.10, 0.8, 0.8, 0.0, 1.0, 1.5),   # more trees
    ('XGB-04', 300, 8, 0.10, 0.8, 0.8, 0.0, 1.0, 1.5),   # deeper trees
    ('XGB-05', 300, 6, 0.05, 0.8, 0.8, 0.0, 1.0, 1.5),   # lower lr
    ('XGB-06', 300, 6, 0.20, 0.8, 0.8, 0.0, 1.0, 1.5),   # higher lr
    ('XGB-07', 300, 6, 0.10, 0.6, 0.6, 0.0, 1.0, 1.5),   # more subsampling
    ('XGB-08', 300, 6, 0.10, 0.8, 0.8, 0.5, 2.0, 1.5),   # L1 + stronger L2
    ('XGB-09', 300, 6, 0.10, 0.8, 0.8, 0.0, 1.0, 1.0),   # Tweedie power=1 (Poisson)
    ('XGB-10', 300, 6, 0.10, 0.8, 0.8, 0.0, 1.0, 1.9),   # Tweedie power=1.9 (near-Gamma)
]

print(f'{len(XGB_VARIANTS)} XGBoost variants defined.')
print(f'Total fits: {len(TWEEDIE_VARIANTS)} × {len(XGB_VARIANTS)} × {len(versions)} versions = '
      f'{len(TWEEDIE_VARIANTS) * len(XGB_VARIANTS) * len(versions)}')
```

## 8 · Preprocessing Helper

```python
def build_features(df: pd.DataFrame, tw_power: float, tw_alpha: float,
                   tw_maxiter: int, xgb_tvp: float) -> tuple:
    """
    Encodes ZONE and JR_SEMN_ACCDN, fits a TweedieRegressor on
    NB_VICTIMES_ZONEHR (clipping to TWEEDIE_CLIP when power > 1),
    appends TWEEDIE_PRED, and returns (X, y, label_encoder).
    """
    df = df.copy()
    df['ZONE'] = zone_le.transform(df['ZONE'])
    df['JR_SEMN_ACCDN'] = df['JR_SEMN_ACCDN'].map(day_map)

    feature_cols = [c for c in df.columns if c not in ('NB_VICTIMES_ZONEHR', 'RISK_CLASS')]

    # Clip target to strictly positive values when Tweedie requires it
    target = df['NB_VICTIMES_ZONEHR'].copy()
    if tw_power > 1:
        target = target.clip(lower=TWEEDIE_CLIP)

    tweedie = TweedieRegressor(power=tw_power, alpha=tw_alpha,
                               solver='lbfgs', max_iter=tw_maxiter)
    tweedie.fit(df[feature_cols], target)
    df['TWEEDIE_PRED'] = tweedie.predict(df[feature_cols])

    le_target = LabelEncoder()
    y = le_target.fit_transform(df['RISK_CLASS'])
    X = df.drop(columns=['NB_VICTIMES_ZONEHR', 'RISK_CLASS'])
    return X, y, le_target
```

## 9 · Main Grid Search Loop

Iterates all 5 versions × 10 Tweedie variants × 10 XGBoost variants = **500 fits**.  
After all combinations are evaluated for a version, the best combination (highest severity score) is identified and its confusion matrix is displayed.

```python
TEST_SIZE = 0.20
all_version_results = {}   # {version: DataFrame of scores}

for ver, df_v in versions.items():
    print(f'\n{"═" * 70}')
    print(f'  VERSION {ver.upper()}')
    print(f'{"═" * 70}')

    # Encode target once to get the split indices — reused across all combos
    df_temp = df_v.copy()
    df_temp['ZONE'] = zone_le.transform(df_temp['ZONE'])
    df_temp['JR_SEMN_ACCDN'] = df_temp['JR_SEMN_ACCDN'].map(day_map)
    le_ref = LabelEncoder()
    y_ref = le_ref.fit_transform(df_temp['RISK_CLASS'])
    idx_train, idx_test = train_test_split(
        np.arange(len(y_ref)), test_size=TEST_SIZE,
        stratify=y_ref, random_state=RANDOM_STATE
    )
    classes = le_ref.classes_
    n_classes = len(classes)

    rows = []
    total = len(TWEEDIE_VARIANTS) * len(XGB_VARIANTS)
    fit_count = 0

    for tw_label, tw_power, tw_alpha, tw_maxiter in TWEEDIE_VARIANTS:

        # Build the full feature matrix for this Tweedie variant
        X_full, y_full, le_fit = build_features(
            df_v, tw_power, tw_alpha, tw_maxiter, tw_power
        )
        X_train = X_full.iloc[idx_train]
        X_test  = X_full.iloc[idx_test]
        y_train = y_full[idx_train]
        y_test  = y_full[idx_test]

        for (xgb_label, n_est, max_d, lr, subsamp, colsamp,
             reg_a, reg_l, xgb_tvp) in XGB_VARIANTS:

            obj = 'multi:softmax' if n_classes > 2 else 'binary:logistic'
            params = dict(
                n_estimators=n_est,
                max_depth=max_d,
                learning_rate=lr,
                subsample=subsamp,
                colsample_bytree=colsamp,
                reg_alpha=reg_a,
                reg_lambda=reg_l,
                tweedie_variance_power=xgb_tvp,
                objective=obj,
                eval_metric='mlogloss',
                n_jobs=-1,
                random_state=RANDOM_STATE,
            )
            if n_classes > 2:
                params['num_class'] = n_classes

            xgb = XGBClassifier(**params)
            xgb.fit(X_train, y_train)
            y_pred = xgb.predict(X_test)

            score = severity_score(y_test, y_pred)
            rows.append({
                'Tweedie': tw_label,
                'tw_power': tw_power,
                'tw_alpha': tw_alpha,
                'XGBoost': xgb_label,
                'n_estimators': n_est,
                'max_depth': max_d,
                'learning_rate': lr,
                'subsample': subsamp,
                'colsample_bytree': colsamp,
                'reg_alpha': reg_a,
                'reg_lambda': reg_l,
                'xgb_tvp': xgb_tvp,
                'Severity Score': score,
                '_y_test': y_test,
                '_y_pred': y_pred,
            })

            fit_count += 1
            if fit_count % 10 == 0:
                print(f'  [{fit_count}/{total}] {tw_label} × {xgb_label} → score: {score:,}')

    results_df = pd.DataFrame(rows)
    best_idx = results_df['Severity Score'].idxmax()
    best = results_df.loc[best_idx]

    print(f'\n  ── Best combination for Version {ver.upper()} ──')
    print(f'  Tweedie : {best["Tweedie"]}  (power={best["tw_power"]}, alpha={best["tw_alpha"]})')
    print(f'  XGBoost : {best["XGBoost"]}  (n_est={best["n_estimators"]}, depth={best["max_depth"]}, lr={best["learning_rate"]})')
    print(f'  Severity Score : {best["Severity Score"]:,}')

    # Confusion matrix for best combination
    cm = confusion_matrix(best['_y_test'], best['_y_pred'])
    fig, ax = plt.subplots(figsize=(6, 5))
    ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classes).plot(
        ax=ax, colorbar=False, cmap='Blues'
    )
    ax.set_title(
        f'Version {ver.upper()} — Best: {best["Tweedie"]} × {best["XGBoost"]}\n'
        f'Severity Score: {best["Severity Score"]:,}',
        fontsize=11, fontweight='bold'
    )
    plt.tight_layout()
    plt.show()

    # Store results without the raw prediction arrays
    all_version_results[ver] = results_df.drop(columns=['_y_test', '_y_pred'])

print('\nAll 500 fits complete.')
```

## 10 · Summary: Best Combination per Version

```python
summary_rows = []
for ver, df_res in all_version_results.items():
    best = df_res.loc[df_res['Severity Score'].idxmax()]
    summary_rows.append({
        'Version': ver.upper(),
        'Tweedie': best['Tweedie'],
        'tw_power': best['tw_power'],
        'tw_alpha': best['tw_alpha'],
        'XGBoost': best['XGBoost'],
        'n_estimators': int(best['n_estimators']),
        'max_depth': int(best['max_depth']),
        'learning_rate': best['learning_rate'],
        'Severity Score': int(best['Severity Score']),
    })

summary_df = pd.DataFrame(summary_rows).set_index('Version')

summary_df.style \
    .background_gradient(subset=['Severity Score'], cmap='RdYlGn') \
    .set_caption('Best Tweedie × XGBoost Combination per Version (highest severity score)')
```

## 11 · Full Score Grid per Version

```python
for ver, df_res in all_version_results.items():
    print(f'\nVersion {ver.upper()} — all 100 combinations (sorted by Severity Score):')
    display_cols = ['Tweedie', 'XGBoost', 'tw_power', 'tw_alpha',
                    'n_estimators', 'max_depth', 'learning_rate', 'Severity Score']
    display(
        df_res[display_cols]
        .sort_values('Severity Score', ascending=False)
        .reset_index(drop=True)
        .style
        .background_gradient(subset=['Severity Score'], cmap='RdYlGn')
        .format({'Severity Score': '{:,}'})
    )
```
